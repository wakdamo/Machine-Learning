```python
# importing classes
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# reading dataset and checking for rows/columns 
dataset= pd.read_csv(r"C:\Users\Amol Wakde\Desktop\Haberman.csv")
dataset.head()
print(dataset.shape)
```

```python
# Labelling the columns/features 
dataset.columns = ['Age','Op_Year','axil_nodes','Surv_status']
print(dataset)

 


```

```python
# To check the high level statistics of the dataset
dataset.describe()
```

```python
# to check the survival more than (Blue Dots) and Less than (Orange Dots)five years 
sns.set_style('white')
sns.FacetGrid(dataset, hue="Surv_status", size=5) \
 .map(plt.scatter, "axil_nodes", "Age") \
 .add_legend();
plt.show();
```

```python
#Pairplots combination to check the density of the data with respect to features
plt.close();
sns.set_style("white");
sns.pairplot(dataset, hue="Surv_status", size=5, vars=['Age','Op_Year', 'axil_nodes'])
plt.show()
```

```python
#Violin plot to check how the data is distributed similat to box plot 
sns.violinplot(x="Surv_status", y="axil_nodes",data=dataset)
plt.legend
plt.show()
```

```python
# to check the survival status with respect to axil_nodes
sns.FacetGrid(dataset,hue="Surv_status", size=8)\
.map(sns.distplot,"axil_nodes")\
.add_legend()
```

```python
#To check the distribution of the data and outliers 
sns.boxplot(x="Surv_status",y="axil_nodes", data=dataset)
plt.show()
```
